## Part Two

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
headers = {
    "User-Agent": "Mozilla/5.0"
}

url = "https://www.lemonde.fr/en/"
response = requests.get(url, headers=headers)
doc = BeautifulSoup(response.text, 'html.parser')

items = doc.find_all(class_='article')

rows = []

for item in items:
    row = {}

    title_tag = item.find(class_='article__title')
    row['title'] = title_tag.text.strip() if title_tag else None

    desc_tag = item.find(class_='article__desc')
    row['subhead'] = desc_tag.text.strip() if desc_tag else None

    try:
        row['url'] = item.select_one('a[href]')['href']
    except:
        row['url'] = None

    row['premium'] = "premium" if item.find(class_='sr-only') else ""

    credit_tag = item.find(class_='article__media-credit')
    row['byline'] = credit_tag.text.strip() if credit_tag else None

    tag_type = item.find(class_='article__category')
    row['type'] = tag_type.text.strip() if tag_type else None

    try:
        img_tag = item.find('img')
        row['image_url'] = img_tag.get('src', None) if img_tag else None
    except:
        row['image_url'] = None

    rows.append(row)

df = pd.DataFrame(rows)
df.to_csv('lemonde_articles.csv', index=False)
print("✅ Saved as lemonde_articles.csv")
```

## Part Three

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

headers = {
    "User-Agent": "Mozilla/5.0"
}

url = "https://travel-id-documents.az.gov/authorized-third-party-driver-license-locations"
response = requests.get(url, headers=headers)

doc = BeautifulSoup(response.text, 'html.parser')

rows = doc.find_all('tr')

data = []

for row in rows[1:]:
    tds = row.find_all('td')
    if len(tds) >= 4:
        name = tds[0].text.strip()
        address = tds[1].text.strip()
        phone = tds[2].text.strip()
        other = tds[3].text.strip()
        
        link_tag = tds[0].find('a')
        link = link_tag.get('href') if link_tag else ''
        
        data.append({
            'Name': name,
            'Address': address,
            'Phone': phone,
            'Other': other,
            'Link': link
        })

df = pd.DataFrame(data)
df.to_csv("az_third_party_locations.csv", index=False)

print("✅ Done. Saved as az_third_party_locations.csv")
```
